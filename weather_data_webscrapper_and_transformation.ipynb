# Web scrapper to extract data from weather website using selenium

### importing libraries

```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import matplotlib.pyplot as plt
from selenium.webdriver.common.by import By

import re
```

### calling webservice

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

### explination of the whole code

```python
#### extracting datafrom webpage
```

```python
## URL of the website from which data is being extracted
year=2016
month=1
yearly_df=pd.DataFrame()
url_temp=f'https://www.wunderground.com/history/monthly/us/fl/miami/KMIA/date/{year}-{month}'
driver.get(url_temp)

### adding sleep time to pause PY-script and load web page completely 
time.sleep(5)


### locating the HTML element containing weather data
temp_vals = driver.find_elements(by=By.CLASS_NAME, value='ng-star-inserted')

### first value contains required data
temp_vals[0].text
```

```python

all_data=temp_vals[0].text

### required data is between ' Pressure (in) Precipitation (in)' and 'Our Apps' using re to extarct data
req_data=re.search("Pressure.{6}Precipitation.*?Our\sApps",all_data, re.DOTALL).group()

## removing trash data
sub_data=re.sub('Pressure.{6}Precipitation.{5}',"",req_data)
sub_data
```

```python
#### above data has format which can be divided into groups based on delimitters between them
result = re.search("^(?P<mnth_name>.{4})(?P<days>.*?\nMax)\sAvg\sMin(?P<temps>.*?\nMax)\sAvg\sMin(?P<dews>.*?\nMax)\sAvg\sMin(?P<humidity>.*?\nMax)\sAvg\sMin(?P<windspeed>.*?\nMax)\sAvg\sMin(?P<pressure>.*?\nTotal)(?P<precip>.*\nOur\sApps)", sub_data,flags=re.DOTALL)
```

```python
# for day number data
dys=result.group('days')
#splitting the data
day_split=re.split(r'\n',dys)
#removing blank values
split_result_filtered = [x for x in day_split if x]
#converting to numbers
days_list=[int(day) for day in split_result_filtered if day.isdigit()]
days_list
```

```python
#for extracting month name
month_name=result.group('mnth_name')
month_name=month_name[1:]
month_list=[month_name]*len(days_list)
month_list
```

```python
# for temp values
temp_vals=result.group('temps')
temp_split=re.split(r'\n',temp_vals)
temp_split_filt=[x for x in temp_split if x]
max_temp=[]
avg_temp=[]
min_temp=[]
#-1 because last value contains max
for temp_set in temp_split_filt[:-1]:
    split_data=temp_set.split()
    max_temp.append(float(split_data[0]))
    avg_temp.append(float(split_data[1]))
    min_temp.append(float(split_data[2]))

temp_split_filt
```

```python
### these are same as above
# for dews values
dew_vals=result.group('dews')
dew_split=re.split(r'\n',dew_vals)
dew_split_filt=[x for x in dew_split if x]
max_dew=[]
avg_dew=[]
min_dew=[]
#-1 because last value contains max
for dew_set in dew_split_filt[:-1]:
    split_data=dew_set.split()
    max_dew.append(float(split_data[0]))
    avg_dew.append(float(split_data[1]))
    min_dew.append(float(split_data[2]))

# for windspeed values
windspeed_vals=result.group('windspeed')
windspeed_split=re.split(r'\n',windspeed_vals)
windspeed_split_filt=[x for x in windspeed_split if x]
max_windspeed=[]
avg_windspeed=[]
min_windspeed=[]
#-1 because last value contains max
for windspeed_set in windspeed_split_filt[:-1]:
    split_data=windspeed_set.split()
    max_windspeed.append(float(split_data[0]))
    avg_windspeed.append(float(split_data[1]))
    min_windspeed.append(float(split_data[2]))


# for pressure values
pressure_vals=result.group('pressure')
pressure_split=re.split(r'\n',pressure_vals)
pressure_split_filt=[x for x in pressure_split if x]
max_pressure=[]
avg_pressure=[]
min_pressure=[]
#-1 because last value contains max
for pressure_set in pressure_split_filt[:-1]:
    split_data=pressure_set.split()
    max_pressure.append(float(split_data[0]))
    avg_pressure.append(float(split_data[1]))
    min_pressure.append(float(split_data[2]))

# for humidity values
humidity_vals=result.group('humidity')
humidity_split=re.split(r'\n',humidity_vals)
humidity_split_filt=[x for x in humidity_split if x]
max_humidity=[]
avg_humidity=[]
min_humidity=[]
#-1 because last value contains max
for humidity_set in humidity_split_filt[:-1]:
    split_data=humidity_set.split()
    max_humidity.append(float(split_data[0]))
    avg_humidity.append(float(split_data[1]))
    min_humidity.append(float(split_data[2]))
```

```python
# for day number data
precip_val=result.group('precip')
#splitting the data
precip_split=re.split(r'\n',precip_val)
#removing blank values
precip_result_filtered = [x for x in precip_split if x]
#converting to numbers
precip_list=[float(precip) for precip in precip_result_filtered[:-1]]

precip_list
```

```python
## combining all data into a Dataframe

combined_data={
            'Month_name':month_list,
            'Days':days_list,
            'Max_temp':max_temp,
            'Avg_temp':avg_temp,
            'Min_temp':min_temp,
            'Max_dew':max_dew,
            'Avg_dew':avg_dew,
            'Min_dew':min_dew,
            'Max_humidity':max_humidity,
            'Avg_humidity':avg_humidity,
            'Min_humidity':min_humidity,
            'Max_windspeed':max_windspeed,
            'Avg_windspeed':avg_windspeed,
            'Min_windspeed':min_windspeed,
            'Max_Pressure':max_pressure,
            'Avg_Pressure':avg_pressure,
            'Min_Pressure':min_pressure,
            'Precipitation_total':precip_list
            
}
        
comb_df=pd.DataFrame(combined_data)
yearly_df=pd.concat([yearly_df,comb_df],axis=0,ignore_index=True)

yearly_df
```

```python
### looping over the years  and months to extract monthly weather data
years=[year for year in range(2012,2024)]
months=[month for month in range(1,13)]
for year in years:
    print('year: ',year)
    yearly_df=pd.DataFrame()
    for month in months:

        ## URL of the website from which data is being extracted
        url_temp=f'https://www.wunderground.com/history/monthly/us/fl/miami/KMIA/date/{year}-{month}'
        driver.get(url_temp)
        
        time.sleep(5)
        
        temp_vals = driver.find_elements(by=By.CLASS_NAME, value='ng-star-inserted')
        
        all_data=temp_vals[0].text
        req_data=re.search("Pressure.{6}Precipitation.*?Our\sApps",all_data, re.DOTALL).group()
        sub_data=re.sub('Pressure.{6}Precipitation.{5}',"",req_data)
        
        
        result = re.search("^(?P<mnth_name>.{4})(?P<days>.*?\nMax)\sAvg\sMin(?P<temps>.*?\nMax)\sAvg\sMin(?P<dews>.*?\nMax)\sAvg\sMin(?P<humidity>.*?\nMax)\sAvg\sMin(?P<windspeed>.*?\nMax)\sAvg\sMin(?P<pressure>.*?\nTotal)(?P<precip>.*\nOur\sApps)", sub_data,flags=re.DOTALL)
        
        
        # for day number data
        dys=result.group('days')
        #splitting the data
        day_split=re.split(r'\n',dys)
        #removing blank values
        split_result_filtered = [x for x in day_split if x]
        #converting to numbers
        days_list=[int(day) for day in split_result_filtered if day.isdigit()]
        
        
        #for extracting month name
        month_name=result.group('mnth_name')
        month_name=month_name[1:]
        month_list=[month_name]*len(days_list)
        
        
        # for temp values
        temp_vals=result.group('temps')
        temp_split=re.split(r'\n',temp_vals)
        temp_split_filt=[x for x in temp_split if x]
        max_temp=[]
        avg_temp=[]
        min_temp=[]
        #-1 because last value contains max
        for temp_set in temp_split_filt[:-1]:
            split_data=temp_set.split()
            max_temp.append(float(split_data[0]))
            avg_temp.append(float(split_data[1]))
            min_temp.append(float(split_data[2]))
        
        
        # for dews values
        dew_vals=result.group('dews')
        dew_split=re.split(r'\n',dew_vals)
        dew_split_filt=[x for x in dew_split if x]
        max_dew=[]
        avg_dew=[]
        min_dew=[]
        #-1 because last value contains max
        for dew_set in dew_split_filt[:-1]:
            split_data=dew_set.split()
            max_dew.append(float(split_data[0]))
            avg_dew.append(float(split_data[1]))
            min_dew.append(float(split_data[2]))
        
        # for windspeed values
        windspeed_vals=result.group('windspeed')
        windspeed_split=re.split(r'\n',windspeed_vals)
        windspeed_split_filt=[x for x in windspeed_split if x]
        max_windspeed=[]
        avg_windspeed=[]
        min_windspeed=[]
        #-1 because last value contains max
        for windspeed_set in windspeed_split_filt[:-1]:
            split_data=windspeed_set.split()
            max_windspeed.append(float(split_data[0]))
            avg_windspeed.append(float(split_data[1]))
            min_windspeed.append(float(split_data[2]))
        
        
        # for pressure values
        pressure_vals=result.group('pressure')
        pressure_split=re.split(r'\n',pressure_vals)
        pressure_split_filt=[x for x in pressure_split if x]
        max_pressure=[]
        avg_pressure=[]
        min_pressure=[]
        #-1 because last value contains max
        for pressure_set in pressure_split_filt[:-1]:
            split_data=pressure_set.split()
            max_pressure.append(float(split_data[0]))
            avg_pressure.append(float(split_data[1]))
            min_pressure.append(float(split_data[2]))
        
        # for humidity values
        humidity_vals=result.group('humidity')
        humidity_split=re.split(r'\n',humidity_vals)
        humidity_split_filt=[x for x in humidity_split if x]
        max_humidity=[]
        avg_humidity=[]
        min_humidity=[]
        #-1 because last value contains max
        for humidity_set in humidity_split_filt[:-1]:
            split_data=humidity_set.split()
            max_humidity.append(float(split_data[0]))
            avg_humidity.append(float(split_data[1]))
            min_humidity.append(float(split_data[2]))
        
        
        # for day number data
        precip_val=result.group('precip')
        #splitting the data
        precip_split=re.split(r'\n',precip_val)
        #removing blank values
        precip_result_filtered = [x for x in precip_split if x]
        #converting to numbers
        precip_list=[float(precip) for precip in precip_result_filtered[:-1]]
        
        combined_data={
            'Month_name':month_list,
            'Days':days_list,
            'Max_temp':max_temp,
            'Avg_temp':avg_temp,
            'Min_temp':min_temp,
            'Max_dew':max_dew,
            'Avg_dew':avg_dew,
            'Min_dew':min_dew,
            'Max_humidity':max_humidity,
            'Avg_humidity':avg_humidity,
            'Min_humidity':min_humidity,
            'Max_windspeed':max_windspeed,
            'Avg_windspeed':avg_windspeed,
            'Min_windspeed':min_windspeed,
            'Max_Pressure':max_pressure,
            'Avg_Pressure':avg_pressure,
            'Min_Pressure':min_pressure,
            'Precipitation_total':precip_list
            
        }
        
        comb_df=pd.DataFrame(combined_data)
        yearly_df=pd.concat([yearly_df,comb_df],axis=0,ignore_index=True)

    file_path=save_path+f"{year}_weather_data.csv"
    yearly_df.to_csv(file_path,index=False)
```
